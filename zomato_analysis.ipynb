```python
df_country= pd.read_excel("Country-Code.xlsx")
```

```python
df.columns
```

```python
#combine these two data frames 

final_df= pd.merge(df, df_country, on= 'Country Code', how= 'left')
final_df
```

```python
# check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names= final_df.Country.value_counts().index
country_names
```

```python
 final_df.Country.value_counts().index
```

```python
 country_val= final_df.Country.value_counts().values
```

```python
#pie chart fo rtop 3 country uses zomato

plt.pie(x=country_val[:3], labels= country_names[:3], autopct='%1.2f%%' )
```

**Observation: Zomato maximum records or transactions are from INDIA tehn USA then from UK**

```python
final_df.columns
```

```python
 ratings= final_df.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

## conclusion : 
1. when rating between 4.5 to 4.9 ----->Excellent
2. when rating between 4.0 to 3.4-------> very good
3. when rating between 3.5 to 3.9 -------> good
4. when rating between 3.0 to 3.4--------> average
5. when rating betweeen 2.5 to 2.9 --------> average
6. when rating between 2.0 to 2.4 ---------> poor

```python
ratings.head()
```

```python
plt.figure(figsize= (12,6))
sns.barplot(x= "Aggregate rating", y = 'Rating Count', data = ratings)
```

```python
plt.figure(figsize= (12,6))
sns.barplot(x= "Aggregate rating", y = 'Rating Count', data = ratings, hue= 'Rating color')
```

Observation :
1. Not rated count is very high
2. Max rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x= "Rating color", data = ratings, hue= 'Rating color')
```
